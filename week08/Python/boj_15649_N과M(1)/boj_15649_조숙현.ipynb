## N과 M(1)

🐾__Q.__ 자연수 N과 M이 주어졌을 때, 아래 조건을 만족하는 <u>길이가 M</u>인 __수열__을 모두 구하는 프로그램을 작성하기
  - 1부터 N까지 자연수 중에서 중복없이 M개를 고른 수열


- 입력
  - 첫째 줄에 자연수 N과 M이 주어짐(1 ≤ M ≤ N ≤ 8)

- 출력
  - 한 줄에 하나씩 조건을 만족하는 수열을 출력
  - 중복되는 수열을 여러 번 출력하면 안되며, 각 수열은 공백으로 구분해서 출력해야 함
  - 수열은 사전 순으로 증가하는 순서로 출력해야 함

```python
# 사전 순으로 출력
# 순열 -> 라이브러리 itertools의 permutations 사용
```

```python
import sys
input = sys.stdin.readline
from itertools import permutations

def permu(N, M):
  ans = list(permutations(range(1, N+1), M))

  for i in range(len(ans)):
    for j in range(M):
      print(ans[i][j], end=' ')
    print()


N, M = map(int, input().split())

permu(N, M)
```

```python
# 재귀로 구현해 보기(DFS)
```

```python
N, M = map(int, input().split())
lst = []

def dfs():

  if len(lst) == M:
    print(' '.join(map(str, lst)))
    return

  else:
    for i in range(1, N+1):
      if i not in lst:
        lst.append(i)
        dfs()
        lst.pop()
```
