```python
import sys
input = sys.stdin.readline
from itertools import permutations

def permu(N, M):
  ans = list(permutations(range(1, N+1), M))

  for i in range(len(ans)):
    for j in range(M):
      print(ans[i][j], end=' ')
    print()


N, M = map(int, input().split())

permu(N, M)
```

```python
N, M = map(int, input().split())
lst = []

def dfs():

  if len(lst) == M:
    print(' '.join(map(str, lst)))
    return

  else:
    for i in range(1, N+1):
      if i not in lst:
        lst.append(i)
        dfs()
        lst.pop()
```
