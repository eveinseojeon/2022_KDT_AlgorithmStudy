## 14888. 연산자 끼워넣기
- __N__개의 수로 이루어진 수열 A1, A2, ..., AN이 주어짐
- 또, 수와 수 사이에 끼워넣을수 있는 __N-1__개의  <u>연산자</u>가 주어짐


- 수와 수 사이에 연산자를 하나씩 넣어서, 수식을 하나 만들 수 있음
  - 이때, 주어진 수의 <u>순서를 바꾸면 안됨</u>

  - 예를 들어
    - 6개의 수로 이루어진 수열이 1, 2, 3, 4, 5, 6이고, 주어진 연산자가 __덧셈(+), 뺄셈(-), 곱셈(x), 나눗셈__으로만 이루어져 있을 경우 __총 60가지__의 식을 만들 수 있음<br>
    
    image.png


- 식의 계산
  - 연산자 우선순위를 무시하고 앞에서 부터 진행해야 함
  - 나눗셈은 __정수 나눗셈 몫만__ 취함
  - 음수를 양수로 나눌 때는 양수로 바꾼 뒤 몫을 취하고, 그 몫을 음수로 바꾼 것과 같음<br>
  image.png

🐾__Q.__ N개의 수와 N-1개의 연산자가 주어졌을 때, 만들 수 있는 식의 결과가 최대인 것과 최소인 것을 구하는 프로그램을 작성하기

  - 입력
    - 첫째 줄에 수의 개수 N(2 ≤ N ≤ 11)가 주어짐
    - 둘째 줄에는 A1, A2, ..., AN이 주어짐 (1 ≤ Ai ≤ 100)
    - 셋째 줄에는 합이 N-1인 4개의 정수가 주어지는데, 차례대로 덧셈(+)의 개수, 뺄셈(-)의 개수, 곱셈(×)의 개수, 나눗셈(÷)의 개수

```python
# N개의 수, N-1개의 연산자가 주어짐
# 연산자 우선순위 무시하고 앞에서 부터 계산
# 나눗셈은 몫만 취함 -> // 
```

```python
N = int(input())
nums_lst = list(map(int, input().split()))
a, s, m, d = map(int, input().split())

max_ans = -1000000001
min_ans = 1000000001

def recursive(cnt, num, add, sub, mul, div):

  global max_ans, min_ans

  if cnt == N:
    max_ans = max(max_ans, num)
    min_ans = min(min_ans, num)
    return

  if add > 0:
    recursive(cnt + 1, num + nums_lst[cnt], add - 1, sub, mul, div)
  if sub > 0:
    recursive(cnt + 1, num - nums_lst[cnt], add, sub - 1, mul, div)
  if mul > 0:
    recursive(cnt + 1, num * nums_lst[cnt], add, sub, mul - 1, div)
  if div > 0:
    recursive(cnt + 1, num // nums_lst[cnt], add, sub, mul, div - 1)


recursive(1, nums_lst[0], a, s, m, d)


# answer
print(max_ans)
print(min_ans)
```
