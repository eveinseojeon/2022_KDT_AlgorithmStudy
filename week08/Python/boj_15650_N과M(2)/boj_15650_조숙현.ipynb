## 15650. N과 M(2)

🐾__Q.__ 자연수 N과 M이 주어졌을 때, 아래 조건을 만족하는 <u>길이가 __M__인 수열</u>을 모두 구하는 프로그램을 작성하기

  - 1부터 __N__까지 자연수 중에서 <u>중복없이</u> __M__개를 고른 수열
  - 고른 수열은 __오른차순__


- 출력
  - 첫째 줄에 자연수 N과 M이 주어짐(1 <= M <= N <= 8)

- 출력
  - 한 줄에 하나씩 문제의 조건을 만족하는 수열을 출력함
  - 중복되는 수열을 여러 번 출력하면 안되며, 각 수열은 공백으로 구분해서 출력해야 함
  - 수열은 사전 순으로 증가하는 순서로 출력해야 함

```python
from itertools import combinations

def combi(N, M):

  ans = list(combinations(range(1, N + 1), M))

  for i in range(len(ans)):
    for j in range(M):
      print(ans[i][j], end=' ')
    print()


N, M = map(int, input().split())

combi(N, M)
```
