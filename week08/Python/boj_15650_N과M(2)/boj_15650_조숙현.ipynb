```python
from itertools import combinations

def combi(N, M):

  ans = list(combinations(range(1, N + 1), M))

  for i in range(len(ans)):
    for j in range(M):
      print(ans[i][j], end=' ')
    print()


N, M = map(int, input().split())

combi(N, M)
```
