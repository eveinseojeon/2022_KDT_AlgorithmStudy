```python
# Íµ¨Í∏Ä Í≤ÄÏÉâÏΩîÎìúÎ•º Ï∞∏Í≥†Ìï®
# ÏôÑÏ†ÑÌûà Ïù¥Ìï¥ Î™ªÌï¥ÏÑú Îã§Ïãú ÌíÄÏñ¥Î≥¥Í∏∞
```

```python
# XÏóêÏÑú Ï∂úÎ∞úÌïòÏó¨ ÏµúÎã®Í±∞Î¶¨KÏóê ÏûàÎäî ÎèÑÏãú Ï∞æÍ∏∞
from collections import deque

N, M, K, X = map(int, input().split()) #ÎèÑÏãú Ïàò, ÎèÑÎ°ú Ïàò, ÏµúÎã®Í±∞Î¶¨, ÏãúÏûëÎèÑÏãú 

graph = [[] for _ in range(N + 1)]
distance = [-1] * (N + 1)
distance[X] = 0 #Ï∂úÎ∞úÎèÑÏãúÎäî 0

for i in range(M):
  a, b = map(int, input().split())
  graph[a].append(b)

def bfs(X): #X:Ï∂úÎ∞úÎèÑÏãú
  queue = deque()
  queue.append(X)

  while queue:
    now = queue.popleft()
    
    for i in graph[now]:  # ÌòÑÏû¨ ÎèÑÏãúÏóêÏÑú Ïù¥ÎèôÌï† Ïàò ÏûàÎäî ÎèÑÏãú ÌôïÏù∏
      if distance[i] == -1:  #Î∞©Î¨∏ÌïòÏßÄ ÏïäÏùÄ ÎèÑÏãúÏù¥Î©¥
        distance[i] = distance[now] + 1   # ÌòÑÏû¨ ÎèÑÏãúÏôÄ Ï∂úÎ∞ú ÎèÑÏãú ÏÇ¨Ïù¥Ïùò Í±∞Î¶¨ +1
        queue.append(i)

  
  check = False  #Ï∂úÎ∞ú ÎèÑÏãúÎ°úÎ∂ÄÌÑ∞ ÏµúÎã®Í±∞Î¶¨ KÏù∏ ÎèÑÏãúÍ∞Ä Ï°¥Ïû¨ÌïòÏßÄ ÏïäÏúºÎ©¥ -1 Ï∂úÎ†•ÏúÑÌï¥, checkÎ≥ÄÏàòÎ•º FalseÎ°ú Ï¥àÍ∏∞Ìôî

  for i in range(1, N + 1):
    if distance[i] == K:
      print(i)
      check = True  # -1Ïù¥ Ï∂úÎ†•ÎêòÏßÄ ÏïäÎèÑÎ°ù Ìï®

  if check == False:
    print(-1)

bfs(X)

```
