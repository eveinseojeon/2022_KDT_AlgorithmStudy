## 2667. 단지번호 붙이기
- <그림 1>과 같이 정사각형 모양의 지도가 있음
  - __1__은 <u>집이 있는 곳</u>을, __0__은 <u>집이 없는 곳</u>을 나타냄
  - 철수는 이 지도를 가지고 연결된 집의 모임인 단지를 정의하고, 단지에 번호를 붙이려 함
  - 여기서 __연결되었다는 것은__ 어떤 집이 <u>좌우, 혹은 아래위로 다른 집이 있는 경우</u>를 의미함
  - 대각선상에 있는 집은 연결된 것이 아님
  - <그림 2>는 <그림 1>을 단지 별로 번호를 붙인 것<br>

- 입력
 - 첫 번째 줄에는 지도의 크기 N(정사각형이므로 가로와 세로의 크기는 같으며 5≤N≤25)이 입력되고, 그 다음 N줄에는 각각 N개의 자료(0혹은 1)가 입력됨

- 출력
  - 첫 번째 줄에는 총 단지수를 출력. 그리고 각 단지내 집의 수를 오름차순으로 정렬하여 한 줄에 하나씩 출력

```python
# 상하좌우 탐색
# n*n배열
# name error 발생
```

```python
# NameError 발생??
from collections import deque

def bfs(graph, a, b):
  dx = [-1, 1, 0, 0]
  dy = [0, 0, -1, 1]

  
  queue = deque()
  queue.append((a, b))
  graph[a][b] = 0  # 탐색한 좌표는 다시 방문하지 않도록 0으로
  cnt = 1

  while queue:
    x, y = queue.popleft()
    for i in range(4): #상하좌우
      nx = x + dx[i]
      ny = y + dy[i]

      global n

      if nx < 0 or nx >= n or ny < 0 or ny >= n:
        continue
      if graph[nx][ny] == 1:
        graph[nx][ny] = 0  # graph의 좌표가 1이면 0으로 바꾸고, cnt에 1씩 더해줌
        queue.append((nx, ny))
        cnt += 1
  return cnt

n = int(input())

graph = [list(map(int, input())) for _ in range(n)]

ans = [] #단지별로 수를 세어 너어줄 리스트
#n*n배열
for i in range(n):
  for j in range(n):
    if graph[i][j] == 1:     #graph의 좌표가 1이면 bfs함수를 실행하여 1의 개수를 세어줌
      ans.append(bfs(graph, i, j))

ans.sort() #오름차 순 정렬

print(len(ans)) #단지의 개수
for i in ans:
  print(i)
```
