### ClimbingStairs
- 계단을 오르는데, n개를 오르면 정상에 도달할 수 있음
- 매번 __1계단__ 또는 __2계단__씩 오를 수 있음<br>
🐾__Q.__ 정상에 오르기 위한 방법은 몇 가지??

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        dp = [1, 2]

        for i in range(2, n + 1):
            dp.append(dp[i - 1] + dp[i - 2])

        return dp[n - 1]
```

```python
import collections

class Solution:
    dp = collections.defaultdict(int)
    
    def climbStairs(self, n: int) -> int:
        
        dp[1] = 1
        dp[2] = 2

        for i in range(3, n+1):

            self.dp[i] = self.climbStairs(i - 1) + self.climbStairs(i - 2)

        return self.dp[n]
```

```python
#타임 리밋
import collections

class Solution:
    dp = collections.defaultdict(int)
    def climbStairs(self, n: int) -> int:
        
        if n <= 2:
            return n

        for i in range(3, n+1):

            self.dp[i] = self.climbStairs(i - 1) + self.climbStairs(i - 2)

        return self.dp[n]
```
