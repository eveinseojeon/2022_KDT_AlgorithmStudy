### FibonacciNumber

```python
class Solution:
    def fib(self, n: int) -> int:
        fibo = [0, 1]

        for i in range(2, n + 1):
            fibo.append(fibo[i - 1] + fibo[i - 2])

        return fibo[n]
```

```python
ans = Solution()
ans.fib(5)
```

```python
import collections

def fibo(n):
  dp = collections.defaultdict(int)
  dp[0] = 0
  dp[1] = 1
  print(dp)

  for i in range(2, n+1):
    dp[i] = dp[i - 1] + dp[i -2]
    print(dp)

  return dp[n]
```

```python
def fibo_2(n: int):
  if n <= 1:
    return n

  else:
    return fibo_2(n-1) + fibo_2(n - 2)
```
