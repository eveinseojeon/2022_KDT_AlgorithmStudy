## Higher Than 75 Marks
__Q.__ STUDENTS테이블에서 75Marks가 넘는 학생의 이름을 조회
- 이름의 뒤에서 3개를 가지고 오름차순 정렬하고, 뒤의 3글자가 같으면, ID로 오름차순 정렬

🐾 부분 문자열 가져오기
- LEFT(컬럼명, n) : 왼쪽을 기준으로 n개의 문자를 가져오는 함수
- RIGHT(컬럼명, n) : 오른쪽을 기준으로 n개의 문자를 가져오는 함수
- MID(컬럼명, 시작위치, n) : 지정한 시작위치를 기준으로 n개의 문자를 가져오는 함수

```python
SELECT NAME FROM STUDENTS
WHERE MARKS > 75
ORDER BY RIGHT(NAME, 3) ASC, ID ASC;
```

## Type of Triangle
__Q.__  세 변의 길이를 사용하여 TRIANGLES 테이블 의 각 레코드 유형 을 식별하는 쿼리를 작성
- 코드에 대해 다음 명령문 중 하나를 출력함
 - Equilateral: It's a triangle with  sides of equal length.
 - Isosceles: It's a triangle with  sides of equal length.
 - Scalene: It's a triangle with  sides of differing lengths.
 - Not A Triangle: The given values of A, B, and C don't form a triangle.
  <br>
  ⏰ 두 변의 합이 한 변보다 작거나 같을 경우 삼각형이 성립되지 않음

- 조건문 사용
  - __CASE WHEN__ 조건문 __THEN__ '표시할 것' __ELSE__ '표시할 것' __END__

```python
SELECT 
CASE 
WHEN A = B AND B = C THEN 'Equilateral'
WHEN A + B <= C OR A + C <= B OR B + C <= A THEN 'Not A Triangle'
WHEN A = B OR A = C OR B = C THEN 'Isosceles'
ELSE 'Scalene'
END
FROM TRIANGLES;
```

## Weather Observation Station 15
__Q.__ STATION테이블에서 LAT_N이 137.2345보다 작고, 그 범위안에서 LAT_N이 가장 큰 LONG_W를 반올림 4한 값을 구하기 

```python
SELECT ROUND(LONG_W, 4) FROM STATION
WHERE LAT_N < 137.2345
ORDER BY LAT_N DESC
LIMIT 1;
```
