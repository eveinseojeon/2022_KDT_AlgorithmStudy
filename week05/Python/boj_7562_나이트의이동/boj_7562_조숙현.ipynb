## 7562. 나이트의 이동
- 체스판 위에 나이트가 놓여져 있음
- 나이트가 한 번에 이동할 수 있는 칸은 아래 그림에 나와있음<BR>
__Q.__ 나이트가 이동하려고 하는 칸이 주어짐. 나이트는 몇 번 움직이면 이 칸으로 이동할 수 있을까??
  - 각 테스트 케이스마다 나이트가 <u>최소 몇 번 만에 이동</u>할 수 있는지 출력 
<br>

image.png


- 입력
  - 첫째 줄: 테스트 케이스의 개수
  - 각 테스트 케이스는 세 줄로 이루어져 있음
    - 첫째 줄: 체스판 한 변의 길이__I__(4<=I<=300)
    - 체스판의 크기는 __I x I__
      - 체스판의 각 칸은 두 수의 쌍 {0, ..., l-1} × {0, ..., l-1}로 나타낼 수 있음
    - 둘째 줄: 나이트가 현재 있는 칸
    - 셋째 줄: 나이트가 이동하려고 하는 칸

- 체스판에서 나이트의 __이동 규칙__
  - <u>상하좌우로 한 칸</u> 움직인 후 <u>꺾어서 두 칸</u> 더 움직인다고 생각하거나 <br><u>상하좌우로 두 칸</u> 움직인 후 <u>꺾어서 한 칸</u> 더 움직인다고 생각할 수 있음
  - 나이트가 움직이는 모습은 알파벳 대문자 __'L'__과 같음

```python
#입력
'''
3
8
0 0
7 0
100
0 0
30 50
10
1 1
1 1
'''
```

```python
from collections import deque

dx = [-1, -1, -2, -2, 1, 1, 2, 2]
dy = [2, -2, 1, -1, 2, -2, 1, -1]

T = int(input())

for i in range(T):
  I = int(input()) #IxI크기의 체스판
  graph = []
  for i in range(I):
    graph.append([0]*I)

  queue = deque()  
  x, y = map(int, input().split()) # 나이트가 현재 있는 칸
  w, z = map(int, input().split()) # 나이트가 이동하려는 칸
  queue.append((x, y))

  while queue:
    x, y = queue.popleft()

    if x == w and y == z:
      break
    
    for j in range(8):
      nx = x + dx[j]
      ny = y + dy[j]
      if nx < 0 or ny < 0 or nx >= I or ny >= I:
        continue
      if graph[nx][ny] == 0:
        graph[nx][ny] = graph[x][y] + 1
        queue.append((nx, ny))

  print(graph[w][z])
```
