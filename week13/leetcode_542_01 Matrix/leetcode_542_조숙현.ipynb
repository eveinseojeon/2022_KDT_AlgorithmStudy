## 01 Matrix

[LeetCode_542](https://leetcode.com/problems/01-matrix/description/)

#### __Q.__ `m x n`의 이진행렬인 `mat`이 주어지고, 해당 셀이 __1__일 때<br>0과의 최단거리를 해당 셀에 표시하여 반환하기

❗
- 리스트 q에 mat을 순회하며 0인 곳의 좌표를 append하고, 1인 곳은 #으로 바꾸어 줌

- mat에서 q에 저장해둔 좌표대로, 값이 0인 부분의 상하좌우를 탐색하여, 범위를 벗어나지 않고 값이 #이면 1을 더해줌
  - 리스트 q에 좌표를 append해 줌

```python
#입력
mat = [[0,0,0],[0,1,0],[1,1,1]]
```

```python
class Solution:
    def updateMatrix(self, mat):
        m = len(mat)      # row
        n = len(mat[0])   # column

        q = []            # 좌표를 저장할 리스트

        for i in range(m):
            for j in range(n):
                if mat[i][j] == 0:
                    q.append((i, j))

                else:
                    mat[i][j] = '#'

        print('값이 0인 부분의 좌표만:', q)
        print()
        print('1인 부분을 #으로:', mat)

        mov=[(1, 0), (-1, 0), (0, 1), (0, -1)]
        for row, col in q:
            for dx, dy in mov:
                m_row = row + dx
                m_col = col + dy

                if 0 <= m_row < m and 0 <= m_col < n and mat[m_row][m_col] == '#':
                    mat[m_row][m_col] = mat[row][col] + 1
                    q.append((m_row, m_col))
        
        print()
        print('탐색한 #의 좌표 append한 리스트 q:', q)
        print()
        
        return mat
```

```python
sol = Solution()
sol.updateMatrix(mat)
```
