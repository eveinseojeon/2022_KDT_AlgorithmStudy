## Minimum Falling Path Sum

[LeetCode_931](https://leetcode.com/problems/minimum-falling-path-sum/submissions/838470858/)


#### __Q.__ 밑으로 떨어지면서 최소값의 합을 구하기

```python
#입력예시
matrix = [[2,1,3],[6,5,4],[7,8,9]]
```

```python
# 틀림
# 문제 자체를 잘못 이해함
#행마다 크기가 가장 작은 값의 위치 인덱스를 저장하는 리스트 만들기
# 그리고 그 인덱스에서 떨어질 수 있는 수들 중에 가장 작은수로 떨어지기
'''
첫번째 행에서 가장 작은 값부터 시작하여 아래로 갈 수 있는 수들 중 가장 작은 값을 더하는 걸로 잘못 이해함
'''

class Solution:
    def minFallingPathSum(self, matrix):
        n = len(matrix)

        if n <= 2:
            ans = []
            for i in range(n):
                m = min(matrix[i])
                ans.append(m)
            
            return sum(ans)

        else:
            down_sum = min(matrix[0])

            min_idx = [0] * n
            min_idx[0] = matrix[0].index(min(matrix[0]))

            for j in range(1, n):
                if min_idx[j-1] == n-1:
                    down_sum += min(matrix[j][min_idx[j-1]], matrix[j][min_idx[j-1]-1])
                    min_idx[j] = matrix[j].index(min(matrix[j][min_idx[j-1]], matrix[j][min_idx[j-1]-1]))

                elif min_idx[j-1] != 0:
                    down_sum += min(matrix[j][min_idx[j-1]-1], matrix[j][min_idx[j-1]], matrix[j][min_idx[j-1]+1])
                    min_idx[j] = matrix[j].index(min(matrix[j][min_idx[j-1]-1], matrix[j][min_idx[j-1]], matrix[j][min_idx[j-1]+1]))

                else:
                    down_sum += min(matrix[j][min_idx[j-1]], matrix[j][min_idx[j-1]+1])
                    min_idx[j] = matrix[j].index(min(matrix[j][min_idx[j-1]], matrix[j][min_idx[j-1]+1]))

            
            return down_sum
```

```python
ans = Solution()
ans.minFallingPathSum(matrix)
```

```python
#구글링 참고

class Solution:
    def minFallingPathSum(self, matrix):
        n = len(matrix)

        if n == 1:
            
            return matrix[0][0]

        else:

            for i in range(1, n):  # 0번째 빼고 1번째 부터 n-1번째 까지
                for j in range(n):
                    if j == n-1:
                        matrix[i][j] = min(matrix[i-1][j], matrix[i-1][j-1]) + matrix[i][j]
                    elif j != 0:
                        matrix[i][j] = min(matrix[i-1][j-1], matrix[i-1][j], matrix[i-1][j+1]) + matrix[i][j]
                    else:
                        matrix[i][j] = min(matrix[i-1][j], matrix[i-1][j+1]) + matrix[i][j]
            
            print(matrix)

            return min(matrix[n-1])
```
