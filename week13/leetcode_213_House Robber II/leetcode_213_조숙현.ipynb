## House Robber II
- 거리의 집들을 도둑질할 예정임
- 각각의 집들마다 도둑질 할 수 있는 돈의 양이 정해져 있음
- 모든 집들은 원형으로 배열되어 있음
  - 첫번째 집과 마지막 집이 이웃되어 있다는 의미
  - 이웃한 집들은 보안시스템이 연결되어 있음
  

[LeetCode_213](https://leetcode.com/problems/house-robber-ii/)

##### ❗__나의 생각__
- nums의 앞, 뒤 집의 보안이 연결되어 있으므로
  1. 도둑이 오른쪽 방향으로 마지막 집은 제외하고 도둑질한 리스트 right를 생성하고,
  2. nums[::-1]로 nums리스트를 뒤집어서 마지막집은 제외하고 도둑질한 리스트를 생성
  3. left, right리스트를 합하여 max()함수로 가장 큰값을 반환하기



```python
class Solution:
    def rob(self, nums):
        if len(nums) <= 3:
            return max(nums)
        
        else:
            #정방향부터
            right = [0] * (len(nums)-1)
            right[0] = nums[0]
            right[1] = max(right[0], nums[1])
            right[2] = max(nums[2] + right[0], right[1])

            #반대 방향부터
            rev_nums = nums[::-1]
            left = [0] * (len(nums))
            left[0] = rev_nums[0]
            left[1] = max(left[0], rev_nums[1])
            left[2] = max(rev_nums[2] + left[0], left[1])

            for i in range(3, len(nums) - 1):
                right[i] = max(nums[i] + right[i-2], right[i - 1])
                left[i] = max(rev_nums[i] + left[i-2], left[i - 1])

            ans = right + left
        
        return max(ans)
```

```python
Sol = Solution()
nums = [1, 2, 3, 1]
nums_2 = [2, 3, 2]
```

```python
Sol.rob(nums)
```

```python
Sol.rob(nums_2)
```
