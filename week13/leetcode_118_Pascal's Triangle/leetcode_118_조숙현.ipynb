## Pascal's Triangle
- __numRows__라는 정수형 입력이 주어짐
- ex) numRows = 5일 경우,

### __Q.__ 위와 같이 numRows가 주어지면, 파스칼의 삼각형을 array로 출력하기

```python
class Solution:
  def generate(self, numRows: int):
    ans = [[1], [1, 1]]
    
    if numRows == 1:
      return [[1]]
    if numRows == 2:
      return ans

    if numRows >= 3:
      for i in range(2, numRows):
        arr = [1, 1]
        for j in range(1, i):
          s = ans[i-1][j-1] + ans[i-1][j]
          arr.insert(j, s)
        
        ans.append(arr)


    return ans
```
