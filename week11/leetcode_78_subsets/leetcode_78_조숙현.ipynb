### __Q.__ nums에서 모든 부분 집합을 리턴하기

```python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        ans = []

        def dfs(idx, sub):
            ans.append(sub)
            
            for i in range(idx, len(nums)):
                dfs(i + 1, sub + [nums[i]])

        dfs(0, [])

        return ans
```

```python
class Solution:
  def subset(self, nums):
    ans = [[]]

    for num in nums:
      ans += [i + [num] for i in ans]

    return ans
```
